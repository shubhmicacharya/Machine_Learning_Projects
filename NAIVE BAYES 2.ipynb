```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("tested.csv")
df
```

```python
df.isna().sum()
```

```python
ab=df.drop(["Age","Cabin","Name","SibSp","Parch","PassengerId","Ticket","Fare","Embarked"], axis='columns')
ab.head(5)
```

```python
target=ab.Survived
features=ab.drop("Survived",axis='columns')
```

```python
target.head(10)
```

```python
features.head(10)
```

```python
dum=pd.get_dummies(features["Sex"])
dum.head(10)
```

```python
new=pd.concat([ab,dum],axis='columns')
new.head(10)
```

```python
final=new.drop(["Sex","Pclass"],axis='columns')
```

```python
final.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(final, target, test_size=0.2, random_state=42)
```

```python
X_train.head(10)
```

```python
y_train.head(10)
```

```python
X_test.head(10)
```

```python
y_test.head(10)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
